```python
def inorder_nodes(root):
    if root:
        yield from inorder_nodes(root.left)
        yield root
        yield from inorder_nodes(root.right)

def reassign_to_bst(root):
    sorted_elements = sorted(node.element for node in inorder_nodes(root))

    for node, element in zip(inorder_nodes(root), sorted_elements):
        node._element = element  # Cheaty encapsulation violation.
```

```python
medium = _Maker(Node).medium()
reassign_to_bst(medium)
draw(medium)
```

```python
import collections

def levels(root):
    if not root:
        return

    cur = [root]

    while cur:
        yield [node.element for node in cur]
        pre = cur
        cur = []
        
        for node in pre:
            if node.left:
                cur.append(node.left)
            if node.right:
                cur.append(node.right)
```
