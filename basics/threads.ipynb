```python
import threading, time
```

```python
def count_up_to_100(): 
    for x in range(100): 
        time.sleep(.05)
        print(x)
```

```python
def print_hello_a_lot(): 
    for _ in range(100): 
        time.sleep(.05)
        print("Hello")
```

```python
t = threading.Thread(target=count_up_to_100)
```

```python
t2 = threading.Thread(target=print_hello_a_lot)
```

```python
t.start()
t2.start()
print("Threads...ATTACK!")
t.join()
t2.join()
```

```python

print("Threads done.")
```
