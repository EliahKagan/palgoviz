```python
from palgoviz.calc import postfix_parse, draw
```

```python
tree = postfix_parse('3 2.2 * 1 + 1 2 / -')
```

```python
draw(tree)
```

```python
tree.evaluate()
```

```python
draw(tree.simplify())
```

```python
tree2 = postfix_parse('3 2.2 * 1 @ 1 2 / -')
```

```python
draw(tree2)
```

```python
tree2_sim = tree2.simplify()
tree2_sim is tree2, tree2_sim.simplify() is tree2_sim
```

```python
draw(tree2_sim)
```

```python
tree2_sim.postfix_serialize()
```

```python
tree2_sim.simplify() is tree2_sim
```
