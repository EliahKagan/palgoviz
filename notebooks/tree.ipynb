```python
root = basic.medium(Node)
subroot = basic.medium(Node).right.left
find_subtree(root, subroot) is find_subtree_fast(root, subroot) is root.right.left
```

```python
find_subtree(subroot, root) is None
```

```python
draw_extended(root, dag=True)  # Here, should look similar to draw/draw_iterative.
```

```python
draw_extended(root, dag=True)  # Now we get deduplication.
```
