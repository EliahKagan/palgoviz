# Inheritance in our `queues` module

```python
from graphviz import Digraph
```

```python
tree = Digraph()
tree.node_attr = dict(shape='none', fontname='Cascadia Mono', fontsize='12')
tree.edge_attr = dict(penwidth='0.5')

tree.edge('Queue', 'FifoQueue')
tree.edge('Queue', 'LifoQueue')
tree.edge('Queue', 'PriorityQueue')
tree.edge('FifoQueue', 'DequeFifoQueue')
tree.edge('FifoQueue', 'AltDequeFifoQueue')
tree.edge('FifoQueue', 'SlowFifoQueue')
tree.edge('FifoQueue', 'BiStackFifoQueue')
tree.edge('FifoQueue', 'SinglyLinkedListFifoQueue')
tree.edge('LifoQueue', 'ListLifoQueue')
tree.edge('LifoQueue', 'DequeLifoQueue')
tree.edge('LifoQueue', 'AltDequeLifoQueue')
tree.edge('LifoQueue', 'SinglyLinkedListLifoQueue')
tree.edge('PriorityQueue', 'FastEnqueueMaxPriorityQueue')
tree.edge('PriorityQueue', 'FastDequeueMaxPriorityQueue')

tree.unflatten(stagger=5)
```
