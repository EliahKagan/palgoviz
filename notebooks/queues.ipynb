## Inheritance in the `queues` module

```python
from graphviz import Digraph
```

```python
tree = Digraph()
tree.node_attr = dict(shape='none', fontname='Cascadia Mono', fontsize='12')
tree.edge_attr = dict(penwidth='0.5')

tree.edge('Queue', 'FifoQueue')
tree.edge('Queue', 'LifoQueue')
tree.edge('Queue', 'PriorityQueue')
tree.edge('FifoQueue', 'DequeFifoQueue')
tree.edge('FifoQueue', 'AltDequeFifoQueue')
tree.edge('FifoQueue', 'SlowFifoQueue')
tree.edge('FifoQueue', 'BiStackFifoQueue')
tree.edge('FifoQueue', 'RingFifoQueue')
tree.edge('RingFifoQueue', 'CompactRingFifoQueue')
tree.edge('FifoQueue', 'SinglyLinkedListFifoQueue')
tree.edge('LifoQueue', 'ListLifoQueue')
tree.edge('LifoQueue', 'DequeLifoQueue')
tree.edge('LifoQueue', 'AltDequeLifoQueue')
tree.edge('LifoQueue', 'SinglyLinkedListLifoQueue')
tree.edge('PriorityQueue', 'FastEnqueueMaxPriorityQueue')
tree.edge('PriorityQueue', 'FastDequeueMaxPriorityQueue')

tree.unflatten(stagger=5)
```

## `SinglyLinkedListLifoQueue` drawings

```python
from palgoviz.queues import SinglyLinkedListLifoQueue
```

```python
lifo1 = SinglyLinkedListLifoQueue()
lifo1.enqueue('Z')
lifo1.enqueue('Y')
lifo1.enqueue('X')

lifo2 = lifo1.copy()
lifo2.dequeue()
lifo2.enqueue('W')

lifo3 = lifo1.copy()
lifo3.enqueue('V')

lifo4 = lifo1.copy()

lifo1.dequeue()
lifo1.dequeue()
lifo1.enqueue('U')
lifo1.enqueue('T')

lifo5 = lifo1.copy()
lifo5.dequeue()
lifo5.dequeue()
lifo5.dequeue()
lifo5.enqueue('S')
lifo5.enqueue('R')

lifo6 = lifo5.copy()
lifo6.enqueue('Q')

SinglyLinkedListLifoQueue.draw(lifo1, lifo2, lifo3, lifo4, lifo5, lifo6)
```
