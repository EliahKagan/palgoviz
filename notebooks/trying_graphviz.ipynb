## Customizing `label` and `shape`

Various drawings of [benzene](https://en.wikipedia.org/wiki/Benzene), to show how to specify custom Graphviz attributes (mainly `label` and `shape`).

Also works on <code>graphviz.***Di***graph</code>.

```python
g = graphviz.Graph()
g.engine = 'circo'  # Usually you'll want to leave the default of 'dot'.

for i in range(6):
    g.node(f'C{i}', label='C')
    g.node(f'H{i}', label='H')
    g.edge(f'C{i}', f'H{i}')

for i in range(6):
    for _ in range(1 + i % 2):
        g.edge(f'C{i}', f'C{(i + 1) % 6}')

g
```

```python
g = graphviz.Graph()
g.engine = 'circo'  # Usually you'll want to leave the default of 'dot'.

for i in range(6):
    g.node(f'C{i}', label='C', shape='none')
    g.node(f'H{i}', label='H', shape='none')
    g.edge(f'C{i}', f'H{i}')

for i in range(6):
    for _ in range(1 + i % 2):
        g.edge(f'C{i}', f'C{(i + 1) % 6}')

g
```

```python
g = graphviz.Graph()
g.engine = 'circo'  # Usually you'll want to leave the default of 'dot'.

for i in range(6):
    g.node(f'C{i}', shape='point')

for i in range(6):
    for _ in range(1 + i % 2):
        g.edge(f'C{i}', f'C{(i + 1) % 6}')

g
```

```python
g = graphviz.Graph()
g.engine = 'circo'  # Usually you'll want to leave the default of 'dot'.

for i in range(6):
    g.node(f'C{i}', label='', shape='circle')
    g.node(f'H{i}', shape='point')
    g.edge(f'C{i}', f'H{i}')

for i in range(6):
    for _ in range(1 + i % 2):
        g.edge(f'C{i}', f'C{(i + 1) % 6}')

g
```
