# The epidemiology of being an umpire

Bob awoke one morning to discover, to his horror, that he was an umpire. He wasn't an umpire at Frank's trial yesterday--but now he was! Before he knew it, he was at a game and people were accepting the things he was screaming at them.

"I've always been a judgmental person," he said. "But how do I know how to make these calls? I didn't even know the rules of baseball yesterday!"

"We're not sure," said the government scientist, whom Bob would soon come to loathe. "But it's a positive turn of events."

"Positive in the sense that my rapid antigen test was positive for me being an umpire?"

"Yes, and also in the sense that America requires a thousandfold increase in umpires."

"A-- What?"

"We need way more baseball games, with fewer people attending each of them. So this domestic epidemic of umpirism is a very fortunate turn of events. Believe me, Bob, we want to know, as much as you do, how this has happened. So we can make it happen way more!"

"Why... why does our country need an explosion of sparsely attended baseball games?"

The government scientist sighed. "Have you not noticed that we are in the middle of a coronavirus pandemic? Baseball is outdoors, and if there are way more games than necessary, attendees will even be spaced far apart. Even accounting for future variants, this is a very safe pastime!"

"So you are trying to fight a coronavirus *pan*demic with an umpirism *epi*demic?"

"Yeah, I don't like it either," the scientist said.
